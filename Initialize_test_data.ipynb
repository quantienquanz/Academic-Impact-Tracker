```python
import numpy as np
import pandas as pd

num_students = 100 

subject_names = [
    'Đại số', 'Giải tích 1', 'Xác suất', 'Thống kê',
    'Thống kê nâng cao', 'Tối ưu hóa', 'Giải tích 2', 'C1', 'C2', 'CTDL',
    'Thuật toán', 'Toán rời rạc', 'Intro AI', 'Học máy', 'Nhập môn DS',
    'Python', 'CC lập trình AI', 'Học sâu', 'Thị giác máy tính',
    'Xử lý ngôn ngữ tự nhiên', 'Khai phá dữ liệu', 'Dữ liệu lớn', 'Thực tập',
    'Kỹ thuật máy tính', 'Cơ sở dữ liệu', 'Mạng máy tính', 'Công nghệ web',
    'Phân tích thiết kế', 'Đồ họa', 'Công nghệ phần mềm',
    'Hệ thống thông tin', 'Kỹ thuật số', 'Lập trình ứng dụng di động'
]

np.random.seed(0)
data = np.random.uniform(4, 10, size=(num_students, len(subject_names)))
msv = np.arange(1, num_students + 1)
names = ['Sinh viên {}'.format(i) for i in msv]
df = pd.DataFrame(data, columns=subject_names)
df.insert(0, 'Họ Tên', names)
df.insert(0, 'MSV', msv)
output_file = 'student_scores.xlsx'
df.to_excel(output_file, index=False)
print(f"Dữ liệu đã được lưu vào tệp Excel: {output_file}")
```
